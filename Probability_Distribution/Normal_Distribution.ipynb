```python
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
```

```python
x_data =np.arange(-5,5,0.00001)
```

```python
y_data =stats.norm.pdf(x_data,0,1)
```

```python
plt.plot(x_data,y_data)
```

```python
z_data=stats.norm.pdf(x_data,0,2) #(x_data,mean,std)
```

```python
plt.plot(x_data,z_data)
```

```python
w_data=stats.norm.pdf(x_data,2,1)
```

```python
plt.plot(x_data,w_data) #Graph gets shifted to 2 units to the right on changing the mean
```
