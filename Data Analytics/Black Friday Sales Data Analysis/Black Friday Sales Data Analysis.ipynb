### Combining Age and Marital Status

```python
lst =[]
for i in range (len(df)):
    lst.append(df["Gender"][i]+"_"+str(df["Marital_Status"][i]))
```

```python
df["MaritalGender"] = lst
```

```python
df.head()
```

```python
# Male Unmarried are purchasing most items in Black Friday Sales
sns.countplot(x=df['MaritalGender'])
```

```python
# Unmarried-Male of age group 26-35 are purchasing most of the products
sns.countplot(x=df['Age'],hue=df['MaritalGender'])
```

```python
sns.countplot(x=df['Product_Category_1'],hue=df['MaritalGender'])
```
