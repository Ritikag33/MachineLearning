### Data Analyis with Multiple Column

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df =pd.read_csv("googleplaystore.csv")
impute = SimpleImputer(missing_values=np.nan,strategy="mean")
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)
df = df.dropna()
```

## GroupBy in Pandas

#### Q5. Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order wrt ratings?

```python
df_pr = df[df["Category"]=="ART_AND_DESIGN"]
df_pr = df_pr[df_pr["Rating"]>4.5]
# Sort values wrt rating in ascending order
df_pr.sort_values(by='Rating',ascending=True)
# In Descending order
df_pr.sort_values(by='Rating',ascending=False).head(5)
```

#### Q6. Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in descending order wrt Reviews

```python
df_pr = df[df["Category"]=='ART_AND_DESIGN']
df_pr = df_pr[df_pr["Rating"]>4.5]
df_pr["Reviews"] = df_pr["Reviews"].astype(int)
df_pr.sort_values(by="Reviews",ascending=False).head(5)
```

#### Q7. Which category is having maximum average rating in descending order?

```python
df['Rating'] = pd.to_numeric(df["Rating"])
# Group by 'Category' and calculate the mean of 'Rating'
df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
```

#### Q8. How many paid apps are there in each category in descending order?

```python
df_pr = df[df["Type"]=='Paid']
df_pr.groupby("Category")["Type"].count().sort_values(ascending=False)
```
