## Matplotlib

Matplotlib is a data visualization library in Python. It provides a high-level interface for drawing attractive and informative statistical graphics. With its wide variety of plotting options, it is used by data scientists and analysts to create visualizations to support data-driven decision making. Matplotlib can be used in various environments, such as Jupyter notebooks, standalone scripts, or web application servers, and supports different output formats, including PNG, PDF, SVG, and JPG.

### 1. Importing matplotlib and some necessary libraries

If we want to plot within an IPython notebook we can use %matplotlib. It comes with two different commands

 1. %matplotlib notebook for interactive plots
 2. %matplotlib inline for static plots
- We will generally use %matplotlib inline

```python
# Used to plot graphs
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
# static plots
%matplotlib inline
```

#### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
x = np.linspace(0,3*np.pi,50)
# Defining the wave
sin_y = np.sin(x)
cos_y = np.cos(x)
# Plotting the wave
plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')
plt.legend(loc= 'upper right')
plt.show()
```

1. Legend is used to add labels to the graph given in the plot function. The loc argument takes an integer or string that specifies the location of the legend on the plot. different values for loc, such as lower right, center, upper right, lower left, etc.
2. title function adds a title to the plot
* '-' will construct a line like in above plots

```python
x = np.linspace(0,3*np.pi,50)
# Defining the wave
sin_y = np.sin(x)
cos_y = np.cos(x)
# Plotting the wave
plt.plot(x,sin_y,'*')
plt.plot(x,cos_y,'--')
plt.title('Sine:* Cosine: - Wave')
plt.show()
```

Note

* In an IPython notebook the graph will be plotted even without using the plt.show() but in some conditions when we are working on big datasets and want to find some inference we use the plt.show() so it is generally a good practice to use it and also if we are writing a python script then also it's necessary to use the plt.show() command

## 4. Different Types of Plots
### 4.1 Line Plot
- A line plot is a type of plot used to visualize the relationship between two continuous variables. It is a plot that displays data points as connected line segments. Line plots are commonly used to visualize trends and patterns in data, especially when the data is continuous and has a large number of observations.

- In a line plot, the x-axis represents one variable and the y-axis represents another variable. The data points are connected by lines to create a plot that shows how the values of the two variables change with respect to each other. Line plots are often used to visualize time-series data, where the x-axis represents time and the y-axis represents a measurement.

- In Matplotlib, the plot function is used to create a line plot. You can specify the x and y values of the data points, along with other optional arguments such as color, line style, and markers . You can also add labels, titles, and legends to your line plot to provide context and clarity.

```python
x = np.linspace(0,4*np.pi,50)
y = np.sin(x)
plt.style.use('Solarize_Light2')
plt.plot(x,y)
plt.title('Sine and Cosine wave')

plt.show()
```

#### 4.1.2 Multiple Line PLots

```python
x = np.linspace(0,4*np.pi,50)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(15,5))
plt.style.use('Solarize_Light2')
plt.plot(x,y_sin,'*',color='black',linewidth=1, label='sin')
plt.plot(x,y_cos,'--',color='blue',linewidth=2,label='cos')
plt.legend()
plt.title('Sine and Cosine wave')

plt.show()
```

#### 4.1.3 For real world datasets

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
target = housing['target']
print('Range:',target.min(),'to',target.max())

plt.figure(figsize=(15,5))
plt.style.use('fast')
# Plotting 1st 200 values
plt.plot(target[:200],color='blue',linewidth=1)
plt.show()
```

#### Linestyle command

```python
x = np.linspace(0,20,21,dtype='int')
plt.figure(figsize=(15,5))
plt.plot(x,x+0,linestyle='solid',label='Solid')
plt.plot(x,x+1,linestyle='dashed',label= 'Dashed')
plt.plot(x,x+2,linestyle='dashdot', label ='Dashdot')
plt.plot(x,x+3,linestyle='dotted',label='Dotted')
plt.legend(fontsize='large')
plt.show()
```

#### 4.1.5 Adjusting the Plot Axes
1. Can be done using plt.xlim for x-axis
2. plt.ylim for y-axis

```python
x = np.linspace(0,4*np.pi,100)
sin_y = np.sin(x)

plt.figure(figsize=(6,2))
# Defining the range of x-axis
plt.xlim([0,10])
# Defining the range of y-axis
plt.ylim([-2,2])
plt.plot(x,sin_y)
plt.title('Sine wave')
plt.show()
```

#### To avoid adjusting graph we can use plt.axis('tight') or plt.axis('equal')

#### 4.1.6 Using .set to combine all of these commands

```python
plt.figure(figsize=(6,2))
x = np.linspace(0,4*np.pi,100)
fig = plt.axes()
fig.plot(x,np.sin(x))
# Can define everything in one go
fig.set(xlim=(0,10),ylim=(-2.5,2.5),xlabel='X-Axes',ylabel='sin(x)',title='Sine wave')
```

### 4.2 Scatter Plots
- A scatter plot is a type of data visualization in which individual data points are represented as points in a two-dimensional plane. The x-coordinate of each data point represents one feature of the data, and the y-coordinate represents another feature. Scatter plots are commonly used to visualize the relationship between two variables and to identify patterns or outliers in the data.

- In Matplotlib, scatter plots can be created using the scatter function. The scatter function accepts two arrays of values, representing the x-coordinates and y-coordinates of the data points. You can also specify additional arguments to customize the appearance of the scatter plot, such as the size and color of the markers, the transparency of the markers, and the marker shape.

```python
# How scattered the data is
plt.figure(figsize=(6,3))
# Uniformly distributed data
n = 1000
x = np.random.randn(n)
y = np.random.randn(n)
# alpha : transparency, s : size of dots
plt.scatter(x,y,c='blue',alpha=.7,s=20)
plt.title('Scatter Plot')
plt.show()
```

In this code, the scatter function is used to create a scatter plot with 500 data points, represented as red markers with a size of 20 and a transparency of 0.7.

#### 4.2..2 Multiple Scatter Plots

```python
plt.figure(figsize=(6,3))
# Uniformly distributed data
n = 1000
x1 = np.random.randn(n)
y1 = np.random.randn(n)
# Shifted 2 pixels towards right
x2 = np.random.randn(n) + 2
y2 = np.random.randn(n) + 2
# alpha : transparency, s : size of dots
plt.scatter(x1,y1,c='darkblue',alpha=.7,s=20,label='Group-1')
plt.scatter(x2,y2,c='darkred',alpha=.7,s=20,label='Group-2')
plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```

### 4.2.3 On Real World Dataset

- xlim ylim axis concepts can be used here as well
1. The primary difference between plt.scatter and plt.plot is that plt.scatter gives us the freedom to control the properties of each individual point (size,face_color,edge_color, transparency level etc)


```python
plt.figure(figsize=(5,4))
rng = np.random.RandomState(0)
x = rng.rand(100)
y = rng.rand(100)
# Defining 100 diff colors and 100 diff sizes
colors = rng.rand(100)
sizes = rng.rand(100)*1000
plt.scatter(x,y,c=colors,s=sizes,alpha=0.7,cmap='viridis',edgecolors='black')
plt.title('Random Scatter Plot')
plt.xlabel('X values')
plt.ylabel('y values')
plt.colorbar()
plt.show()
```

* Alpha: Transparency of the markers in the scatter plot.
* Colorbar: A bar showing the colors used in the plot and the corresponding values.
* Colormap: A color mapping used to convert scalar data to colors.
* Edgecolors: The color of the edges of the markers in a scatter plot.
* Sizes: The size of the markers in a scatter plot.

### 4.3 Histograms and Binnings
A histogram is a plot that displays the distribution of a set of continuous or discrete data. The data is divided into a set of intervals or bins, and the height of each bar in the plot represents the number of data points that fall within that bin. Binnings refers to the process of dividing the data into these bins. The choice of bin size can greatly affect the appearance of the histogram, so it's important to choose a bin size that accurately represents the underlying distribution of the data.

```python
plt.figure(figsize=(5,3))
data = np.random.randn(1000)
plt.hist(data,edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(5,3))
data = np.random.randn(10000000)
# Change the number of bins
plt.hist(data,edgecolor='white',color='green',alpha=0.9,bins=2000)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title("Random Histogram")
plt.show()
```

Individual Bins can be removed using histtype='stepfilled'



```python
plt.figure(figsize=(5,3))
data = np.random.randn(1000)
# Change the number of bins
plt.hist(data,edgecolor='black',color='red',alpha=0.7,bins=20,histtype='stepfilled')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title("Random Histogram")
plt.show()
```

#### 4.3.2 Multiple Histograms

```python
plt.figure(figsize=(5,3))
x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 1
x3 = np.random.randn(1000) + 2
# Change the number of bins
plt.hist(x1,alpha=0.7,bins=20,edgecolor='black')
plt.hist(x2,alpha=0.7,bins=20,edgecolor='black')
plt.hist(x3,alpha=0.7,bins=20,edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title("Random Histogram")
plt.show()
```

### 4.4 2D Histograms
plt.hist2d is a function in matplotlib that generates a 2D histogram. A 2D histogram is a representation of the distribution of data over two variables. In contrast to a regular histogram which shows the distribution of a single variable, a 2D histogram shows the relationship between two variables. The function creates a heatmap that shows the density of data points in a 2D space.

```python
plt.figure(figsize=(5,3))
x = np.random.randn(1000)
y = np.random.randn(1000) 
# Change the number of bins
plt.hist2d(x,y,bins=20,cmap='inferno')
plt.xlabel('x')
plt.ylabel('y')
plt.title("2D Histogram")
plt.show()
```

```python
plt.figure(figsize=(5,3))
x = np.random.randn(1000)
y = np.random.randn(1000) 
# Change the number of bins
plt.hist2d(x,y,bins=20,cmap='Blues')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.title("2D Histogram")
plt.show()
```

### 4.5 Pie Charts
A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

```python
cars = ['AUDI','BMW','TESLA','FORD','JAGUAR','MERCEDES']
data = [450,675,3,35,200,600]
fig = plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,wedgeprops=dict(edgecolor='black'))
plt.show()
```

#### 4.5.2 Basic Operations

```python
cars = ['AUDI','BMW','TESLA','FORD','JAGUAR','MERCEDES']
data = [450,675,20,55,200,600]
# Exploding
explode = [0,0,.2,.1,0,0]
fig = plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,wedgeprops=dict(edgecolor='black'),explode=explode)
plt.show()
```

#### Adding shadows

```python
cars = ['AUDI','BMW','TESLA','FORD','JAGUAR','MERCEDES']
data = [450,675,20,55,200,600]
# Exploding
explode = [0,0,.2,.1,0,0]
fig = plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,wedgeprops=dict(edgecolor='black'),explode=explode,shadow=True)
plt.show()
```

#### Adding Percentage

```python
cars = ['AUDI','BMW','TESLA','FORD','JAGUAR','MERCEDES']
data = [450,675,40,65,200,600]
# Exploding
explode = [0,0,.2,.1,0,0]
fig = plt.figure(figsize=(7,7))
# autopct: how many places of decimal you want
plt.pie(data,labels=cars,wedgeprops=dict(edgecolor='black'),explode=explode,shadow=True,autopct='%2.2f%%')
plt.show()
```

#### Adding Widget

```python
fig = plt.figure(figsize=(7,7))
# autopct: how many places of decimal you want
plt.pie(data,labels=cars,wedgeprops={'edgecolor':'white','linewidth':2,'antialiased':True},shadow=True,autopct='%2.2f%%')
plt.show()
```

```python
# Blur Edges
fig = plt.figure(figsize=(7,7))
# autopct: how many places of decimal you want
plt.pie(data,labels=cars,wedgeprops={'edgecolor':'white','linewidth':2,'antialiased':False},shadow=True,autopct='%2.2f%%')
plt.show()
```

### 4.6 Multiple Subplots
Multiple subplots refer to dividing a figure into multiple subplots or axes, each of which can contain a separate plot. This allows for multiple plots to be displayed in a single figure. In matplotlib, you can create multiple subplots using the subplot function.

- Index Based

#### Index Based

```python
x = np.linspace(0,10,50)
fig, ax = plt.subplots(2,2)
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sine Wave')

y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')


y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')

y = np.random.randn(1000)
ax[1,1].hist(y,bins=40,edgecolor='black')
ax[1,1].set_title('Random Distribution')

plt.tight_layout()
plt.show()
```

### With loops

```python
plt.subplots_adjust(hspace=.4,wspace=.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(.5,.5,str((2,3,i)),fontsize=15,ha='center')
    plt.tight_layout()
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)
for i in range(1,2):
    fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(6*3,6))
    
    ax1.plot(x,y)
    ax2.plot(x,y**2)
    ax3.plot(x,y**3)
    
    ax1.set_title('Sine Wave')
    ax2.set_title('Sine Wave - square')
    ax3.set_title('Sine Wave - cube')
    
    plt.tight_layout()
    plt.show()
```
