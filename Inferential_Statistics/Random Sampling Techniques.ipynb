# Stratified Sampling

```python
df['origin'].unique()
```

```python
#Group by origin and get 5 members of each unique("origin")
df.groupby("origin",group_keys=False).apply(lambda x:x.sample(5))
```

```python
df["origin"].value_counts()
```

```python
proportionate_stratified_sampling=df.groupby("origin",group_keys=False).apply(lambda x:x.sample(frac=0.1))
```

```python
proportionate_stratified_sampling["origin"].value_counts()
```
